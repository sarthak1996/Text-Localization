```python
import numpy as np
import os
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from PIL import Image
```

```python
from skimage import color
from skimage import io
```

```python
max_x=0
max_y=0
max_x_index=1
max_y_index=1
```

```python
for index in range(1,411):
    img_name="img_"+str(index)
    img_loc="Train_Gray_Images/"
    extension=".jpg"
    #Read image
    if(not(os.path.isfile(img_loc+img_name+extension))):
        extension=".png"
    if(not(os.path.isfile(img_loc+img_name+extension))):
        extension=".gif"
    img = io.imread(img_loc+img_name+extension);
    if(img.shape[0]>max_x):
        max_x=img.shape[0]
        max_x_index=index
    if(img.shape[1]>max_y):
        max_y=img.shape[1]
        max_y_index=index
```

```python
def plotMaxes(max_x,max_x_index,max_y,max_y_index):
    img_name="img_"+str(max_x_index)
    img_loc="Train_Gray_Images/"
    extension=".jpg"
    if(not(os.path.isfile(img_loc+img_name+extension))):
        extension=".png"
    if(not(os.path.isfile(img_loc+img_name+extension))):
        extension=".gif"
    img = io.imread(img_loc+img_name+extension);
    plt.imshow(img)
    plt.show()
    
    img_name="img_"+str(max_y_index)
    img_loc="Train_Gray_Images/"
    extension=".jpg"
    if(not(os.path.isfile(img_loc+img_name+extension))):
        extension=".png"
    if(not(os.path.isfile(img_loc+img_name+extension))):
        extension=".gif"
    img = io.imread(img_loc+img_name+extension);
    plt.imshow(img)
    plt.show()
```

```python
plotMaxes(max_x,max_x_index,max_y,max_y_index)
```
